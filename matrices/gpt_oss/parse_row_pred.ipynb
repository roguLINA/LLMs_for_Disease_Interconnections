```python
ALL_ICD_CODES = sorted(simple_icd_10.get_all_codes(with_dots=False))

def expand_icd_range(codes_list):
    expanded_list = []
    for item in codes_list:
        # Проверяем, является ли элемент диапазоном (напр. "A01-A09")
        if isinstance(item, str) and '-' in item:
            match = re.match(r'^([A-Z][0-9]{2})-([A-Z][0-9]{2})$', item)
            if match:
                start_code, end_code = match.groups()
                # Находим все коды из справочника, попадающие в диапазон
                # Важно: сравниваем только коды категорий (длиной 3 символа)
                in_range = [
                    code for code in ALL_ICD_CODES 
                    if len(code) == 3 and start_code <= code <= end_code
                ]
                expanded_list.extend(in_range)
                continue
        
        # Если это обычный код, просто добавляем его
        expanded_list.append(item)
    return expanded_list
```

```python
df['B'] = df['B'].apply(expand_icd_range)
df['C'] = df['C'].apply(expand_icd_range)
```
